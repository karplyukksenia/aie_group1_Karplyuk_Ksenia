Чанкинг документов

Для разбиения документов реализован простой чанкинг по словам с перекрытием.

**Выбранные параметры:**
- `chunk_size = 25` слов
- `chunk_overlap = 5` слов

**Обоснование:**
- Документы короткие (максимум 50 слов), поэтому большинство разобьются на 2 чанка с перекрытием.
- Чанкинг по словам позволяет избежать обрезания слов посередине и выглядит естественнее, чем чанкинг по символам.
- Параметры выбраны так, чтобы продемонстрировать работу overlap на практике.

```python
def simple_chunk_text(text: str, chunk_size: int = 25, chunk_overlap: int = 5) -> list[str]:
    if not text:
        return []

    words = text.split()
    chunks = []
    start = 0
    n_words = len(words)

    while start < n_words:
        end = start + chunk_size
        chunk_words = words[start:end]
        chunk = " ".join(chunk_words)
        chunks.append(chunk)

        if end >= n_words:
            break

        start = end - chunk_overlap

    return chunks
```

```python
chunked_data = []

for doc in knowledge_base:
    chunks = simple_chunk_text(doc["text"], chunk_size=25, chunk_overlap=5)
    for i, chunk_text in enumerate(chunks):
        chunked_data.append({
            "doc_id": doc["id"],
            "title": doc["title"],
            "chunk_id": f"{doc['id']}_c{i}",
            "text": chunk_text,
            "chunk_length": len(chunk_text)
        })

print(f"Всего создано чанков: {len(chunked_data)}")
print(f"Средняя длина чанка: {np.mean([c['chunk_length'] for c in chunked_data]):.1f} символов")
print(f"Мин. длина чанка: {min(c['chunk_length'] for c in chunked_data)} | Макс. длина: {max(c['chunk_length'] for c in chunked_data)}")
```

### Эмбеддинги и индекс FAISS

Для получения векторных представлений текстовых фрагментов выбрана модель эмбеддингов  
**`sentence-transformers/all-MiniLM-L6-v2`**.

Обоснование выбора:
- Это одна из самых популярных и эффективных лёгких моделей для семантического поиска
- Размер эмбеддинга = 384 (компактный и быстрый)
- Хорошо работает с английским языком

Будет использовано FAISS с индексом `IndexFlatIP` (Inner Product), который после нормализации векторов эквивалентен косинусному сходству

```python
!pip install -q sentence-transformers
```

```python
from sentence_transformers import SentenceTransformer

model_name = "sentence-transformers/all-MiniLM-L6-v2"
embedding_model = SentenceTransformer(model_name, device=device)

print(f"Модель эмбеддингов загружена: {model_name}")
print(f"Размерность эмбеддинга: {embedding_model.get_sentence_embedding_dimension()}")
print(f"Устройство: {device}")
```

```python
chunk_texts = [chunk["text"] for chunk in chunked_data]

embeddings = embedding_model.encode(
    chunk_texts,
    convert_to_numpy=True,
    normalize_embeddings=True,
    show_progress_bar=True
)

print(f"Эмбеддинги получены. Форма: {embeddings.shape}")
```

```python
dimension = embeddings.shape[1]

index = faiss.IndexFlatIP(dimension)
index.add(embeddings)

print(f"FAISS индекс создан успешно")
print(f"Размер индекса: {index.ntotal} векторов")
```

```python
def search_faiss(query: str, top_k: int = 5):
    query_embedding = embedding_model.encode(
        [query],
        convert_to_numpy=True,
        normalize_embeddings=True
    )


    scores, indices = index.search(query_embedding, top_k)

    results = []
    for rank, (score, idx) in enumerate(zip(scores[0], indices[0]), 1):
        chunk = chunked_data[idx]
        results.append({
            "rank": rank,
            "score": float(score),
            "doc_id": chunk["doc_id"],
            "title": chunk["title"],
            "text": chunk["text"],
            "chunk_id": chunk["chunk_id"]
        })

    return results
```

```python
test_queries = [
    "What is a list in Python?",
    "How to add elements to a dictionary?",
    "What is the difference between list and tuple?",
    "Explain lambda functions",
    "How to handle errors in Python?"
]

print("Демонстрация поиска")

for i, query in enumerate(test_queries, 1):
    print(f"\nЗапрос {i}: '{query}'")
    results = search_faiss(query, top_k=3)

    for res in results:
        print(f"  {res['rank']}. [{res['title']}]  score={res['score']:.4f}")
        print(f"     {res['text'][:180]}..." if len(res['text']) > 180 else f"     {res['text']}")
    print()
```

### Небольшой эксперимент с параметрами retrieval

Будет проведён эксперимент по влиянию размера чанка на качество retrieval.

Сравниваемые варианты:
- Вариант A (базовый): `chunk_size = 25` слов, `chunk_overlap = 5` слов
- Вариант B: `chunk_size = 5` слов, `chunk_overlap = 1` слово

Обоснование:

При уменьшении размера чанка до 5 слов значительно возрастает количество фрагментов (гранулярность поиска растёт), но при этом теряется контекст внутри каждого чанка.  
Эксперимент позволяет оценить, что важнее в данной базе знаний — контекст или высокая гранулярность.

```python
def evaluate_with_new_chunking(chunk_size: int, chunk_overlap: int, top_k: int = 3):
    chunked_data_new = []
    for doc in knowledge_base:
        chunks = simple_chunk_text(doc["text"], chunk_size=chunk_size, chunk_overlap=chunk_overlap)
        for i, chunk_text in enumerate(chunks):
            chunked_data_new.append({
                "doc_id": doc["id"],
                "title": doc["title"],
                "chunk_id": f"{doc['id']}_c{i}",
                "text": chunk_text
            })

    chunk_texts = [c["text"] for c in chunked_data_new]
    embeddings_new = embedding_model.encode(
        chunk_texts,
        convert_to_numpy=True,
        normalize_embeddings=True,
        show_progress_bar=False
    )

    dim = embeddings_new.shape[1]
    index_new = faiss.IndexFlatIP(dim)
    index_new.add(embeddings_new)

    hits_at_5 = []
    for item in control_queries:
        query = item["query"]
        expected_id = item["expected_doc_id"]

        query_emb = embedding_model.encode([query], convert_to_numpy=True, normalize_embeddings=True)
        _, indices = index_new.search(query_emb, top_k)

        retrieved_ids = [chunked_data_new[idx]["doc_id"] for idx in indices[0]]
        hit = 1 if expected_id in retrieved_ids else 0
        hits_at_5.append(hit)

    hit_rate = sum(hits_at_5) / len(hits_at_5)
    num_chunks = len(chunked_data_new)

    return hit_rate, num_chunks
```

```python
print("Эксперимент: влияние размера чанка")
print()


hit_a, chunks_a = evaluate_with_new_chunking(chunk_size=25, chunk_overlap=5)
hit_b, chunks_b = evaluate_with_new_chunking(chunk_size=5, chunk_overlap=1)

print(f"Вариант A (chunk_size=25 слов, overlap=5)   →  чанков: {chunks_a:2d}   |  Hit@3 = {hit_a:.3f}")
print(f"Вариант B (chunk_size=5 слов, overlap=1)    →  чанков: {chunks_b:2d}   |  Hit@3 = {hit_b:.3f}")
```

Уменьшение размера чанка до 5 слов сильно увеличивает гранулярность поиска (количество чанков выросло с 54 до 257), но при этом каждый чанк содержит очень мало контекста

Эксперимент позволяет увидеть, что важно сохранять достаточный контекст внутри чанка для успешного retrieval

При уменьшении размера чанка с 25 до 5 слов количество чанков выросло почти в 5 раз (54 → 257), но качество retrieval немного упало (Hit@3 с 1.000 до 0.900). Это показывает, что в данной базе знаний (короткие документы) избыточно мелкий чанкинг ухудшает качество, потому что теряется необходимый контекст внутри фрагмента. Лучше использовать умеренный размер чанка (20–30 слов)

```python
chunked_data_extended = []

for doc in knowledge_base_extended:
    chunks = simple_chunk_text(doc["text"], chunk_size=25, chunk_overlap=5)
    for i, chunk_text in enumerate(chunks):
        chunked_data_extended.append({
            "doc_id": doc["id"],
            "title": doc["title"],
            "chunk_id": f"{doc['id']}_c{i}",
            "text": chunk_text
        })

chunk_texts_extended = [chunk["text"] for chunk in chunked_data_extended]
embeddings_extended = embedding_model.encode(
    chunk_texts_extended,
    convert_to_numpy=True,
    normalize_embeddings=True,
    show_progress_bar=True
)

dimension = embeddings_extended.shape[1]
index_extended = faiss.IndexFlatIP(dimension)
index_extended.add(embeddings_extended)

print(f"Новый индекс содержит {index_extended.ntotal} векторов")
```

```python
comparison_queries = [
    {
        "query": "What is a set in Python and why is it useful?",
        "expected_after": "doc26"   # новый документ
    },
    {
        "query": "How to perform union and intersection on collections in Python?",
        "expected_after": "doc27"
    },
    {
        "query": "When should I use a set instead of a list?",
        "expected_after": "doc28"
    }
]

comparison_results = []

for item in comparison_queries:
    q = item["query"]

    # в старом индексе
    results_before = search_faiss(q, top_k=3)
    retrieved_before = [r["doc_id"] for r in results_before]

    # в новом индексе
    def search_extended(query: str, top_k: int = 3):
        query_emb = embedding_model.encode([query], convert_to_numpy=True, normalize_embeddings=True)
        scores, indices = index_extended.search(query_emb, top_k)
        results = []
        for rank, (score, idx) in enumerate(zip(scores[0], indices[0]), 1):
            chunk = chunked_data_extended[idx]
            results.append({
                "rank": rank,
                "score": float(score),
                "doc_id": chunk["doc_id"],
                "title": chunk["title"],
                "text": chunk["text"]
            })
        return results

    results_after = search_extended(q, top_k=3)
    retrieved_after = [r["doc_id"] for r in results_after]

    changed = retrieved_before != retrieved_after[:len(retrieved_before)]

    comparison_results.append({
        "query": q,
        "before_retrieved_sources": ", ".join(retrieved_before),
        "after_retrieved_sources": ", ".join(retrieved_after),
        "changed": changed
    })

comp_df = pd.DataFrame(comparison_results)
comp_df.to_csv("artifacts/retrieval_before_after_update.csv", index=False, encoding="utf-8")

display(comp_df)
```

После добавления трёх новых документов о множествах (`set`) в Python поведение системы retrieval изменилось:

- Запросы, связанные с `set`, которые раньше возвращали в основном документы про списки и словари, теперь корректно находят новые документы на первых позициях
- Во всех трёх тестовых запросах состав retrieved_sources изменился (`changed = True`)
- Это подтверждает, что добавление новых знаний + переиндексация работают корректно и позволяют RAG-системе использовать свежую информацию


```python
def mini_rag(query: str, top_k: int = 3):
    results = search_faiss(query, top_k=top_k)

    context_parts = []
    sources = []

    for res in results:
        context_parts.append(
            f"Источник: {res['title']}\n"
            f"{res['text']}"
        )
        sources.append(res['title'])

    context = "\n\n".join(context_parts)

    answer = (f"{context}\n")

    return answer.strip(), sources
```

```python
rag_examples = [
    {
        "question": "What is a list in Python and how is it created?",
        "answer": None,
        "retrieved_sources": None
    },
    {
        "question": "What methods can I use to add or remove elements from a list?",
        "answer": None,
        "retrieved_sources": None
    },
    {
        "question": "What is the difference between mutable and immutable objects?",
        "answer": None,
        "retrieved_sources": None
    },
    {
        "question": "What is a set in Python and when should I use it?",   # новый документ
        "answer": None,
        "retrieved_sources": None
    },
    {
        "question": "How can I handle exceptions and errors in Python?",
        "answer": None,
        "retrieved_sources": None
    }
]


for ex in rag_examples:
    answer, sources = mini_rag(ex["question"])
    ex["answer"] = answer
    ex["retrieved_sources"] = ", ".join(sources)
    print(f"Вопрос: {ex['question']}")
    print(f"Источники: {ex['retrieved_sources']}")
    print()
    print(f"Ответ: {answer[:300]}...\n")
    print('--------')
```
